<h1 style="color:blue">Building collaborative filtering model from scratch</h1>
We will recommend items based on user-user similarity and item-item similarity. For that, first we need to calculate the number of unique users and movies.

```python
grp_user_items.head(50)
```

```python
try:
    for i in grp_user_items:
        grp_user_items[i] = pd.to_numeric(grp_user_items[i])
except:
    pass
```

```python
data = grp_user_items.drop(["customer_name", "item_name"], axis=1)
data.head()
```

```python
n_users = data["customer_id"].unique().max()
n_items = data["item_id"].unique().max()
print(f"n_users {n_users}\tn_items {n_items}")
```

<h3 style="color:#88292f">Now, we will create a user-item matrix which can be used to calculate the similarity between users and items.</h3>

```python
data_matrix = np.zeros((n_users, n_items))
for line in data.itertuples():
    data_matrix[line[1]-1, line[2]-1] = line[3]

# print(data_matrix.shape)
# data_matrix = data_matrix[~np.all(data_matrix == 0, axis=1)]
print(data_matrix.shape)
```

<h3 style="color:#88292f">Now, we will calculate the similarity. We can use the pairwise_distance function from sklearn to calculate the cosine similarity.</h3>

```python
from sklearn.metrics.pairwise import pairwise_distances

user_similarity = pairwise_distances(data_matrix, metric="cosine")
item_similarity = pairwise_distances(data_matrix.T, metric="cosine")
```

```python
user_similarity.shape
```

<h3 style="color:#88292f">This gives us the item-item and user-user similarity in an array form. The next step is to make predictions based on these similarities. Letâ€™s define a function to do just that.</h3>

```python
def predict(freq_count, similarity, type="user"):
    if type == 'user':
        mean_user_freq_cout = freq_count.mean(axis=1)
        
        #We use np.newaxis so that mean_user_freq_count has same format as freq_item_count
        freq_count_diff = (freq_count - mean_user_freq_cout[:, np.newaxis])
        pred = mean_user_freq_cout[:, np.newaxis] + similarity.dot(freq_count_diff) / np.array([np.abs(similarity).sum(axis=1)]).T
        
    elif type == "item":
        pred = freq_count.dot(similarity) / np.array([np.abs(similarity).sum(axis=1)])
        
    return pred


user_prediction = predict(data_matrix, user_similarity, "user")
item_prediction = predict(data_matrix, item_similarity, "item")
```

```python
output_df = pd.DataFrame(columns=['customer_id', "customer_name", "item_id", "item_name", "similarity_score"])

for customer_id, similarity in enumerate(user_prediction):
    for item_id, val in enumerate(similarity):
        try:
            customer_name = grp_user_items["customer_name"].where(grp_user_items["customer_id"] == customer_id + 1).dropna().values[0]
            item_name = grp_user_items["item_name"].where(grp_user_items["item_id"] == item_id + 1).dropna().values[0]
            output_df = output_df.append({"customer_id": customer_id + 1, "customer_name": customer_name, "item_id": item_id + 1, "item_name": item_name, "similarity_score": val}, ignore_index=True)

        except IndexError:
            pass
    
```

<h1 style="color:blue">Building a recommendation engine using matrix factorization</h1>

```python
class MF:
    # Initializing the user-item freq matrix, no. of latent features, alpha and beta.
    def __init__(self, R, K, alpha, beta, iterations):
        self.R = R
        self.num_users, self.num_items = R.shape
        self.K = K
        self.alpha = alpha
        self.beta = beta
        self.iterations = iterations
        
    
    # intializing user features and item features matrix
    def train(self):
        self.P = np.random.normal(scale=1./self.K, size=(self.num_users, self.K))
        self.Q = np.random.normal(scale=1./self.K, size=(self.num_items, self.K))
        
        # intializing the bias terms
        self.b_u = np.zeros(self.num_users)
        self.b_i = np.zeros(self.num_items)
        self.b = np.mean(self.R[np.where(self.R != 0)])
        
        
        # List of feq samples
        self.samples = [(i, j, self.R[i, j]) for i in range(self.num_users) for j in range(self.num_items) if self.R[i, j] > 0]
        
        
        # stochastic gradient  decent for given number of iteration
        training_process = []
        for i in range(self.iterations):
            np.random.shuffle(self.samples)
            self.sgd()
            mse = self.mse()
            
            training_process.append((i, mse))
            if (1 + i) % 20 == 0:
                print("Iteration. %d ; error = %.4f" % (i + 1, mse))
                
        return training_process
    
    
    # computing total mean squared error
    def mse(self):
        xs, ys = self.R.nonzero()
        predicted = self.full_matrix()
        
        error = 0
        for x, y in zip(xs, ys):
            error += pow(self.R[x, y] - predicted[x, y], 2)
            
        return np.sqrt(error)
    
    
    # Stochastic gradient descent to get optimized P and Q matrix
    def sgd(self):
        for i, j, r in self.samples:
            prediction = self.get_freq(i, j)
            e = (r - prediction)
            
            self.b_u[i] += self.alpha * (e - self.beta * self.b_u[i])
            self.b_i[j] += self.alpha * (e - self.beta * self.b_i[j])
            
            self.P[i, :] += self.alpha * (e * self.Q[j, :] - self.beta * self.P[i, :])
            self.Q[j: :] += self.alpha * (e * self.P[i, :] - self.beta * self.Q[j, :])
            
            
    # freq for user i and item j
    def get_freq(self, i, j):
        prediction = self.b + self.b_u[i] + self.b_i[j] + self.P[i, :].dot(self.Q[j, :].T)
        return prediction
    
    
    # Full user-item freq matrix
    def full_matrix(self):
        return self.b + self.b_u[:, np.newaxis] + self.b_i[np.newaxis:,] + self.P.dot(self.Q.T)
```
