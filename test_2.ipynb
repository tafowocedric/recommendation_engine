## Top-N accuracy metrics

<p>This evaluation method works as follows:</p>

<ul>
    <li>For each user:
        <ul>
            <li>For each item the user has interacted in test set:
                <ul>
                    <li>Sample 100 other items the user has never interacted.
    Ps. Here we naively assume those non interacted items are not relevant to the user, which might not be true, as the user may simply not be aware of those not interacted items. But let's keep this assumption.</li>
                    <li>Ask the recommender model to produce a ranked list of recommended items, from a set composed one interacted item and the 100 non-interacted ("non-relevant!) items</li>
                    <li>Compute the Top-N accuracy metrics for this user and interacted item from the recommendations ranked list</li>
                </ul>
            </li>
        </ul>
    </li>
    <li>Aggregate the global Top-N accuracy metrics</li>
<\ul>

```python
#Indexing by customer_id to speed up the searches during evaluation
interactions_full_indexed_df = interactions_full_df.set_index('customer_id')
interactions_train_indexed_df = interactions_train_df.set_index('customer_id')
interactions_test_indexed_df = interactions_test_df.set_index('customer_id')

interactions_full_indexed_df
```

```python
"lenght of all available items %d" % len(grp_user_items["item_id"].unique())
```

```python
def get_items_interacted(customer_id, interactions_df):
    # get the user's data and merge in the item information
    interacted_items = interactions_df.loc[customer_id]["item_id"]
    return set(interacted_items if type(interacted_items) == pd.Series else [interacted_items])
```

```python
# Top-N accuracy metrics consts
EVAL_RANDOM_SAMPLE_NON_INTERACTED_ITEMS = 5

class ModelEvaluator:
    def get_not_interacted_items_sample(self, customer_id, sample_size, seed=42):
        interacted_items = get_items_interacted(customer_id=customer_id, interactions_df=interactions_full_indexed_df)
        
        all_items = set(grp_user_items["item_id"])
        non_interacted_items = all_items - interacted_items
        
        random.seed(seed)
        non_interacted_items_sample = random.sample(non_interacted_items, sample_size)
        
        return set(non_interacted_items_sample)
    
    
    def _verify_hit_top_n(self, item_id, recommended_items, topn):
        try:
            index = next(i for i, c in enumerate(recommended_items) if c == item_id)
        except:
            index = -1
            
        hit = int(index in range(0, topn))
        return hit, index
    
    
    def evaluate_model_for_user(self, model, customer_id):
        # getting the items in test set
        interacted_values_testset = interactions_test_indexed_df.loc[customer_id]
        
        if type(interacted_values_testset["item_id"]) == pd.Series:
            customer_interacted_items_testset = set(interacted_values_testset["item_id"])
        else:
            customer_interacted_items_testset = set([interacted_values_testset["item_id"]])
            
        interacted_items_count_testset = len(customer_interacted_items_testset)
        
        # getting a ranked recommendation list from a model for a given customer
        customer_recs_df = model.recommend_items(customer_id=customer_id, items_to_ignore=get_items_interacted(customer_id, interactions_df=interactions_train_indexed_df), topn=6)
        
        hits_at_3_count = 0
        hits_at_5_count = 0
        # For each item the user has interacted in test set
        for item_id in customer_interacted_items_testset:
            # getting random samples 5 items the customer has not interacted
            # to represent items that are assumed to be non revelant to the customer
            non_interacted_items_sample = self.get_not_interacted_items_sample(customer_id, sample_size=EVAL_RANDOM_SAMPLE_NON_INTERACTED_ITEMS, seed=item_id%(2**32))
            
            # combining the current interacted item with the 5 random items
            items_to_filter_recs = non_interacted_items_sample.union(set([item_id]))
            
            # Filtering only recommendations that are either the interacted item or from a random sample of 5 non-interacted items
            valid_recs_df = customer_recs_df[customer_recs_df["item_id"].isin(items_to_filter_recs)]
            valid_recs = valid_recs_df["item_id"].values
            
            # Verifying if the current interacted item is among the Top-N recommended items
            hits_at_3, index_at_3 = self._verify_hit_top_n(item_id, valid_recs, 3)
            hits_at_3_count += hits_at_3
            
            hits_at_5, index_at_10 = self._verify_hit_top_n(item_id, valid_recs, 5)
            hits_at_5_count += hits_at_5
            
            
        # Recall is the rate of interacted items that are ranked among the Top-N recommended items
        # when mixed with a set of non-relevant items
        recall_at_3 = hits_at_3_count / float(interacted_items_count_testset)
        recall_at_5 = hits_at_5_count / float(interacted_items_count_testset)
        
        customer_metrics = {
            "hits@3_count": hits_at_3_count,
            "hits@5_count": hits_at_5_count,
            "interacted_count": interacted_items_count_testset,
            "recall@3": recall_at_3,
            "recall@5": recall_at_5
        }
        return customer_metrics
    
    
    def evaluate_model(self, model):
        customers_metrics = []
        for idx, customer_id in enumerate(list(interactions_test_indexed_df.index.unique().values)):
            customer_metrics = self.evaluate_model_for_user(model=model, customer_id=customer_id)
            customer_metrics["_customer_id"] = customer_id
            customers_metrics.append(customer_metrics)

        detailed_results_df = pd.DataFrame(customers_metrics).sort_values("interacted_count", ascending=False)
        
        global_recall_at_3 = detailed_results_df["hits@3_count"].sum() / float(detailed_results_df["interacted_count"].sum())
        global_recall_at_5 = detailed_results_df["hits@5_count"].sum() / float(detailed_results_df["interacted_count"].sum())
        
        global_metrics = {
            "ModelName": model.get_model_name(),
            "recall@3": global_recall_at_3,
            "recall@5": global_recall_at_5
        }
        
        return global_metrics, detailed_results_df
        
        
model_evaluator = ModelEvaluator()
```

# Popularity Model
A common (and usually hard-to-beat) baseline approach is the Popularity model. This model is not actually personalized - it simply recommends to a user the most popular items that the user has not previously consumed. As the popularity accounts for the "wisdom of the crowds", it usually provides good recommendations, generally interesting for most people.
Ps. The main objective of a recommender system is to leverage the long-tail items to the users with very specific interests, which goes far beyond this simple technique.

```python
# computes the most popular items
item_popularity_df = interactions_full_df.groupby(["item_id"])["freq_item_count"].sum().sort_values(ascending=False).reset_index()
item_popularity_df["freq_item_count"] = item_popularity_df["freq_item_count"] / 10
item_popularity_df
```

```python
class PopularityRecommender:
    MODEL_NAME = "Popularity"
    
    def __init__(self, popularity_df, items_df=None):
        self.popularity_df = popularity_df
        self.items_df = items_df
        
    def get_model_name(self):
        return self.MODEL_NAME
    
    def recommend_items(self, customer_id, items_to_ignore=[], topn=10, verbose=False):
        # recommend the more popular items that the user hasn't seen yet.
        recommendations_df = self.popularity_df[~self.popularity_df["item_id"].isin(items_to_ignore)].sort_values("freq_item_count", ascending=False).head(topn)
        if  verbose:
            if self.items_df is None:
                raise Exception('"items_df" is required in verbose mode')
            self.items_df.drop("freq_item_count", axis=1, inplace=True)
            recommendations_df = recommendations_df.merge(self.items_df, how="left", left_on="item_id", right_on="item_id")[["item_id", "item_name", "description", "freq_item_count"]]
            
        return recommendations_df
    
# test
popularity_model = PopularityRecommender(item_popularity_df, interactions_from_selected_users_df)
popularity_model

print("Evaluating Popularity Recommendation Model...")
pop_global_metrics, pop_detailed_results_df = model_evaluator.evaluate_model(popularity_model)
print("\nGlobal metrics:\n%s" % pop_global_metrics)
pop_detailed_results_df
```

# Content-Based Filtering model
Content-based filtering approaches leverage description or attributes from items the user has interacted to recommend similar items. It depends only on the user previous choices, making this method robust to avoid the cold-start problem.

Here we are using a very popular technique in information retrieval (search engines) named TF-IDF. This technique converts unstructured text into a vector structure, where each word is represented by a position in the vector, and the value measures how relevant a given word is for an item. As all items will be represented in the same Vector Space Model, it is to compute similarity between items.

```python
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler, normalize

import scipy
```

```python
interactions_from_selected_users_df
```

```python
# ignoring stopwords (words with no sematics) from English
stop_words = stopwords.words("english")

# train a model whose vectors size is 5000, composed by the main unigrams and bigrams found in the corpus ignoring stopwords
vectorizer = TfidfVectorizer(analyzer="word", max_features=24, stop_words=stop_words)

item_ids = interactions_from_selected_users_df["item_id"].tolist()
tfidf_matrix = vectorizer.fit_transform(interactions_from_selected_users_df["item_name"] + " " + interactions_from_selected_users_df["description"])

tfidf_feature_name = vectorizer.get_feature_names()
tfidf_matrix
```

```python
set(item_ids)
```

To model the customer profile, we take all the item profiles the customer has interacted and average them. The average is weighted by the interaction strength, in other words, the items the customer has interacted the most will have a higher strength in the final customer profile.

```python
def get_item_profile(item_id):
    idx = item_ids.index(item_id)
    item_profile = tfidf_matrix[idx: idx + 1]
    return item_profile


def get_item_profiles(ids):
    if type(ids) != pd.Series:
        ids = [ids]
        
    item_profiles_list = [get_item_profile(x) for x in ids]
    item_profiles = scipy.sparse.vstack(item_profiles_list)
    
    return item_profiles


def build_users_profile(customer_id, interactions_indexed_df):
    interactions_customer_df = interactions_indexed_df.loc[customer_id]
    customer_item_profiles = get_item_profiles(interactions_customer_df["item_id"])
    
    customer_item_strengths = np.array(interactions_customer_df["freq_item_count"]).reshape(-1, 1)
    
    # weighted average of item profiles by the interaction strength
    customer_item_strengths_weighted_avg = np.sum(customer_item_profiles.multiply(customer_item_strengths), axis=0) / np.sum(customer_item_strengths)
    customer_profile_norm = normalize(customer_item_strengths_weighted_avg)
    
    return customer_profile_norm
    
    
def build_customer_profiles():
    interactions_indexed_df = interactions_train_df[interactions_train_df["item_id"].isin(grp_user_items["item_id"])].set_index("customer_id")
    
    user_profiles = {}
    for customer_id in interactions_indexed_df.index.unique():
        user_profiles[customer_id] = build_users_profile(customer_id, interactions_indexed_df)
        
    return user_profiles

customer_profiles = build_customer_profiles()
print("User Len ", len(customer_profiles))
```

```python
myprofile = customer_profiles[1]
print(myprofile.shape)
pd.DataFrame(sorted(zip(tfidf_feature_name, customer_profiles[1].flatten().tolist()), key=lambda x: -x[1]), columns=["token", "relevance"]).head(50)
```

```python
class ContentBasedRecommender:
    MODEL_NAME = 'Content-Based'
    
    def __init__(self, items_df=None):
        self.item_ids = item_ids
        self.items_df = items_df
        
    def get_model_name(self):
        return self.MODEL_NAME
    
    def _get_similar_items_to_customers_profile(self, custormer_id, topn=10):
        #Computes the cosine similarity between the customer profile and all item profiles
        cosine_similarities = cosine_similarity(customer_profiles[custormer_id], tfidf_matrix)
        
        # Get the top similar items
        similar_indices = cosine_similarities.argsort().flatten()
        
        # Sort the similar items by similarity
        similar_items = sorted([(item_ids[i], cosine_similarities[0, i]) for i in similar_indices], key=lambda x: -x[1])
        return similar_items
    
    def recommend_items(self, customer_id, items_to_ignore=[], topn=10, verbose=False):
        similar_items = self._get_similar_items_to_customers_profile(customer_id)
        
        # ignores items the user already interacted
        similar_items_filtered = list(filter(lambda x: x[0] not in items_to_ignore, similar_items))
        
        recommendations_df = pd.DataFrame(similar_items_filtered, columns=["item_id", "recStrength"]).drop_duplicates().head(topn)

        if  verbose:
            if self.items_df is None:
                raise Exception('"items_df" is required in verbose mode')
            
            recommendations_df = recommendations_df.merge(self.items_df, how="left", left_on="item_id", right_on="item_id")[["item_id", "item_name", "description", "recStrength"]]
            
        return recommendations_df
    
# test
content_based_recommender_model = ContentBasedRecommender(interactions_from_selected_users_df)

print('Evaluating Content-Based Filtering model...')
cb_global_metrics, cb_detailed_results_df = model_evaluator.evaluate_model(content_based_recommender_model)
print('\nGlobal metrics:\n%s' % cb_global_metrics)
cb_detailed_results_df
```

# Collaborative Filtering model
Collaborative Filtering (CF) has two main implementation strategies:

<ul>
    <li>
        <strong>Memory-based:</strong> This approach uses the memory of previous users interactions to compute users similarities based on items they've interacted (user-based approach) or compute items similarities based on the users that have interacted with them (item-based approach).
A typical example of this approach is User Neighbourhood-based CF, in which the top-N similar users (usually computed using Pearson correlation) for a user are selected and used to recommend items those similar users liked, but the current user have not interacted yet. This approach is very simple to implement, but usually do not scale well for many users. A nice Python implementation of this approach in available in <a href="http://muricoca.github.io/crab/">Crab</a>.
    </li>
    <li>
        <strong>Model-based:</strong> This approach, models are developed using different machine learning algorithms to recommend items to users. There are many model-based CF algorithms, like neural networks, bayesian networks, clustering models, and latent factor models such as Singular Value Decomposition (SVD) and, probabilistic latent semantic analysis.
    </li>
</ul>

## Matrix Factorization
Latent factor models compress user-item matrix into a low-dimensional representation in terms of latent factors. One advantage of using this approach is that instead of having a high dimensional matrix containing abundant number of missing values we will be dealing with a much smaller matrix in lower-dimensional space.

```python
from scipy.sparse import csr_matrix
from scipy.sparse.linalg import svds
```

```python
users_items_pivot_sparse_matrix = csr_matrix(users_items_pivot_matrix)
users_items_pivot_sparse_matrix
```

```python
# The number of factors to factor the user-item matrix.
NUMBER_OF_FACTORS_MF = 5

# performs matrix factorization of the original user item matrix
U, sigma, Vt = svds(users_items_pivot_sparse_matrix, k=NUMBER_OF_FACTORS_MF)
sigma = np.diag(sigma)

print(f"U.shape {U.shape}\t Vt.shape {Vt.shape}\tsigma.shape {sigma.shape}")
```

```python
all_users_predicted_ratings = np.dot(np.dot(U, sigma), Vt)
all_users_predicted_ratings.shape, all_users_predicted_ratings
```

After the factorization, we try to reconstruct the original matrix by multiplying its factors. The resulting matrix is not sparse any more. It was generated predictions for items the user have not yet interaction, which we will exploit for recommendations.

```python
all_user_predicted_ratings_norm = (
    all_users_predicted_ratings - all_users_predicted_ratings.min()) \
    / (all_users_predicted_ratings.max() - all_users_predicted_ratings.min())

all_user_predicted_ratings_norm
```

```python
# converting the reconstructed matrix back to a pandas dataframe
cf_preds_df = pd.DataFrame(all_user_predicted_ratings_norm, columns=users_items_pivot_matrix_df.columns, index=users_ids).transpose()
cf_preds_df
```

```python
len(cf_preds_df.columns)
```

```python
class CFRecommender:
    MODEL_NAME = "Collaborative"
    
    def __init__(self, cf_predictions_df, items_df=None):
        self.cf_predictions_df = cf_predictions_df
        self.items_df = items_df
        
    def get_model_name(self):
        return self.MODEL_NAME
    
    def recommend_items(self, customer_id, items_to_ignore=[], topn=10, verbose=False):
        # get and sort the customer predictions
        sorted_customer_predictions = self.cf_predictions_df[customer_id].sort_values(ascending=False).reset_index().rename(columns={customer_id: "recStrength"})
        
        # recommend the highest predicted rating items that the user hasn't seen yet
        recommendations_df = sorted_customer_predictions[~sorted_customer_predictions["item_id"].isin(items_to_ignore)].sort_values("recStrength", ascending=False).drop_duplicates().head(topn)
        
        if verbose:
            if self.items_df is None:
                raise Exception('"items_df" is required in verbose mode')
                
            recommendations_df = recommendations_df.merge(self.items_df, how="left", left_on="item_id", right_on="item_id")[["item_id", "item_name", "description", "recStrength"]]
            
        return recommendations_df
    

cf_recommender_model = CFRecommender(cf_preds_df, grp_user_items)
```

```python
print('Evaluating Collaborative Filtering (SVD Matrix Factorization) model...')
cf_global_metrics, cf_detailed_results_df = model_evaluator.evaluate_model(cf_recommender_model)
print('\nGlobal metrics:\n%s' % cf_global_metrics)
cf_detailed_results_df
```

# Hybrid Recommender
What if we combine Collaborative Filtering and Content-Based Filtering approaches?
Would that provide us with more accurate recommendations?
In fact, hybrid methods have performed better than individual approaches in many studies and have being extensively used by researchers and practioners.
Let's build a simple hybridization method, as an ensemble that takes the weighted average of the normalized CF scores with the Content-Based scores, and ranking by resulting score. In this case, as the CF model is much more accurate than the CB model, the weights for the CF and CB models are 100.0 and 1.0, respectivelly.

```python
class HybridRecommender:
    MODEL_NAME = "Hybrid"
    
    def __init__(self, cb_rec_model, cf_rec_model, items_df, cb_ensemble_weigth=1.0, cf_ensemble_weight=1.0):
        self.cb_rec_model = cb_rec_model
        self.cf_rec_model = cf_rec_model
        self.items_df = items_df
        self.cb_ensemble_weigth = cb_ensemble_weigth
        self.cf_ensemble_weight = cf_ensemble_weight
        
    def get_model_name(self):
        return self.MODEL_NAME
    
    def recommend_items(self, customer_id, items_to_ignore=[], topn=10, verbose=False):
        # Getting the top-10 content-based filtering recommendations
        cb_recs_df = self.cb_rec_model.recommend_items(customer_id, items_to_ignore, topn, verbose).rename(columns={"recStrength": "recStrengthCB"}).drop_duplicates()
        
        # Getting the top-10 collaborative filtering recommendations
        cf_recs_df = self.cf_rec_model.recommend_items(customer_id, items_to_ignore, topn, verbose).rename(columns={"recStrength": "recStrengthCF"}).drop_duplicates()
        
        # Combining the results by item_id
        if verbose:
            cb_recs_df.drop(["description", "item_name"], axis=1, inplace=True)
            
        recs_df = cb_recs_df.merge(cf_recs_df, how="outer", left_on="item_id", right_on="item_id").fillna(0.0)
        
        #Computing a hybrid recommendation score based on CF and CB scores
        #recs_df['recStrengthHybrid'] = recs_df['recStrengthCB'] * recs_df['recStrengthCF'] 
        recs_df["recStrengthHybird"] = (recs_df["recStrengthCB"] * self.cb_ensemble_weigth) + (recs_df["recStrengthCF"] * self.cf_ensemble_weight)

        # Sorting recommendation by hybrid score
        recommendations_df = recs_df.sort_values("recStrengthHybird", ascending=False).head(topn)
        if verbose:
            if self.items_df is None:
                raise Exception('"items_df" is required in verbose mode')
                
            recommendations_df.drop(["description", "item_name"], axis=1, inplace=True)
            recommendations_df = recommendations_df.merge(self.items_df, how="left", left_on="item_id", right_on="item_id")[["item_id", "item_name", "description", "recStrengthHybird"]]
            
        return recommendations_df
    
# test
hybrid_recommender_model = HybridRecommender(content_based_recommender_model, cf_recommender_model, interactions_from_selected_users_df, cb_ensemble_weigth=1.0, cf_ensemble_weight=100.0)

print('Evaluating Hybrid model...')
hybrid_global_metrics, hybrid_detailed_results_df = model_evaluator.evaluate_model(hybrid_recommender_model)
print('\nGlobal metrics:\n%s' % hybrid_global_metrics)
hybrid_detailed_results_df
```
