## Loading Data And Feature Engineering

```python
business_id = 1

sales_items_table = pd.read_sql_query(f"SELECT item_id, item_name, item_type, sales_id FROM user_sales_items WHERE business_id='{business_id}'", con=engine)
sales_items_table.set_index("sales_id", inplace=True)

user_sales_table = pd.read_sql_query(f"SELECT id, customer_id, customer_name  FROM user_sales_table WHERE business_id='{business_id}'", con=engine)
user_sales_table.set_index("id", inplace=True)

user_items_sales = pd.merge(user_sales_table, sales_items_table, how="outer", left_index=True, right_index=True, validate="many_to_many").dropna()
user_items_sales.set_index("item_id", inplace=True)

items = pd.read_sql_query(f"SELECT id, Description FROM qb_item WHERE business_id='{business_id}'", con=engine)
items.set_index("id", inplace=True)

user_items_sales = pd.merge(user_items_sales, items, how="outer", left_index=True, right_index=True, validate="many_to_many").dropna()
user_items_sales.index.name = "item_id"
user_items_sales.reset_index(inplace=True)

user_items_sales.columns = user_items_sales.columns.str.lower()
user_items_sales.tail(50)
```

## recommendation system => user cold-start problem
<p>Recommender systems have a problem known as user cold-start, in which is hard do provide personalized recommendations for users with none or a very few number of consumed items, due to the lack of information to model their preferences.</p>
<p>For this reason, we are keeping in the dataset only users with at leas 2| interactions.</p>

```python
users_interactions_count_df = grp_user_items.groupby(["customer_id", "item_id"]).size().groupby("customer_id").size()
print("# users: %d" % len(users_interactions_count_df))

users_with_enough_interactions_df = users_interactions_count_df[users_interactions_count_df > 1].reset_index()[['customer_id']]
print('# users with at least 2 interactions: %d' % len(users_with_enough_interactions_df))
```

```python
print('# of interactions: %d' % len(grp_user_items))
interactions_from_selected_users_df = grp_user_items.merge(users_with_enough_interactions_df,
                                                         how="right",
                                                         left_on="customer_id",
                                                         right_on="customer_id")
print('# of interactions from users with at least 2 interactions: %d' % len(interactions_from_selected_users_df))
```

```python
interactions_from_selected_users_df
```

<p>To model the user interest on a given article, we aggregate all the interactions the user has performed in an item by a weighted sum of interaction type strength and apply a log transformation to smooth the distribution.</p>

```python
def smooth_user_preference(x):
    return math.log(1 + x, 2)


interactions_full_df = interactions_from_selected_users_df.groupby(["customer_id", "item_id"])["freq_item_count"].sum()\
    .apply(smooth_user_preference).reset_index()

print("# of unique user/item  interactions %d" % len(interactions_full_df))
interactions_full_df
```

```python
n_factors = 5
x_train_array = [x_train[:, 0], x_train[:, 1]]
x_test_array = [x_test[:, 0], x_test[:, 1]]
```

```python
import tensorflow as tf

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Reshape, Dot, Embedding, Add, Activation, Lambda
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2
```

```python
def RecommenderV1(n_customers, n_items, n_factors):
    user = Input(shape=(1,))
    u = Embedding(n_customers, n_factors, embeddings_initializer="he_normal", embeddings_regularizer=l2(1e-6))(user)
    u = Reshape((n_factors,))(u)
    
    item = Input(shape=(1,))
    i = Embedding(n_items, n_factors, embeddings_initializer="he_normal", embeddings_regularizer=l2(1e-6))(item)
    i = Reshape((n_factors,))(i)
    
    x = Dot(axes=1)([u, i])
    
    model = Model(inputs=[user, item], outputs=x)
    opt = Adam(lr=0.001)
    model.compile(loss="mean_squared_error", optimizer=opt)
    
    return model


model = RecommenderV1(n_customer, n_items, n_factors)
model.summary()
```

```python
history = model.fit(x=x_train_array, y=y_train, batch_size=5, epochs=15, verbose=1, validation_data=(x_test_array, y_test))
```

```python
class EmbeddingLayer:
    def __init__(self, n_items, n_factors):
        self.n_items = n_items
        self.n_factors = n_factors
        
    def __call__(self, x):
        x = Embedding(self.n_items, self.n_factors, embeddings_initializer="he_normal", embeddings_regularizer=l2(1e-6))(x)
        x = Reshape((self.n_factors, ))(x)
        
        return x
    
def RecommenderV2(n_customers, n_items, n_factors, min_freq_count, max_freq_count):
    user = Input(shape=(1,))
    u = EmbeddingLayer(n_customers, n_factors)(user)
    ub = EmbeddingLayer(n_customers, 1)(user)

    item = Input(shape=(1,))
    i = EmbeddingLayer(n_items, n_factors)(item)
    ib = EmbeddingLayer(n_items, 1)(item)

    x = Dot(axes=1)([u, i])
    x = Add()([x, ub, ib])
    x = Activation("sigmoid")(x)
    x = Lambda(lambda x: x * (max_freq_count - min_freq_count) + min_freq_count)(x)

    model = Model(inputs=[user, item], outputs=x)
    opt = Adam(lr=0.001)
    model.compile(loss="mean_squared_error", optimizer=opt)

    return model


model = RecommenderV2(n_customer, n_items, n_factors, min_freq_count, max_freq_count)
model.summary()
```

```python
history = model.fit(x=x_train_array, y=y_train, batch_size=5, epochs=15, verbose=1, validation_data=(x_test_array, y_test))
```

```python
from tensorflow.keras.layers import Concatenate, Dense, Dropout

def RecommenderNet(n_customers, n_items, n_factors, min_freq_count, max_freq_count):
    user = Input(shape=(1,))
    u = EmbeddingLayer(n_customers, n_factors)(user)
    
    item = Input(shape=(1,))
    i = EmbeddingLayer(n_items, n_factors)(item)
    
    x = Concatenate()([u, i])
    x = Dropout(0.05)(x)
    
    x = Dense(10, kernel_initializer='he_normal')(x)
    x = Activation('relu')(x)
    x = Dropout(0.5)(x)
    
    x = Dense(1, kernel_initializer='he_normal')(x)
    x = Activation('sigmoid')(x)
    x = Lambda(lambda x: x * (max_freq_count - min_freq_count) + min_freq_count)(x)
    
    model = Model(inputs=[user, item], outputs=x)
    opt = Adam(lr=0.001)
    model.compile(loss='mean_squared_error', optimizer=opt)
    
    return model


model = RecommenderNet(n_customer, n_items, n_factors, min_freq_count, max_freq_count)
model.summary()
```

```python
history = model.fit(x=x_train_array, y=y_train, batch_size=5, epochs=15, verbose=1, validation_data=(x_test_array, y_test))
```
